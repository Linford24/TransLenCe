```python
Variables
  - A variable is a custom name you give to any data type in Python for easy referencing.
  - You have already gone through the basic Python data types namely, numbers(integer and float), strings, boolean.
```

```python
- Assuming you have this statement,
     "I am the best programmer in the world. I build great solutions!"
      This statement is a string since its enclosed in quotes.
      In case you want to send this statement to your 3 best friends, rewriting it each time would be tedious and boring.
      Variables remove the need to rewrite the statement repeatedly 3 times. 
      It allows you to write it just once and send it to all your friends.

    
```

```python
# So we can give the statement a variable name.
motivation = "I am the best programmer in the world. I build great solutions!"
```

```python
# Now, when sending the message to your friends, you just need to send motivation
# and when they click on motivation, they will see the full message, "I am the best programmer in the world. I build great solutions!" display.

motivation
```

```python
- Likewise,
  writing a this float, 2333.6883903926527829203 every single time is boring,
  so you can assign it to a variable. 
```

```python
#Let's call the variable, my_float.
my_float = 2333.6883903926527829203
```

```python
# Now writing my_float then returns the number stored in it
my_float
```

```python
Hint:

Variables can be created in variety of syntax.
In the examples above, the "snake_case" syntax was used

The snake case syntax uses just lowercase for naming the variables. 
If the variable name contains multiple words like, school and group, you put an underscore between forming, school_group.

Likewise for flatcase,
 school_group becomes schoolgroup.
 You just combine both words without any separator.

For camelCase,
  school_group becomes schoolGroup

For PascalCase
  school_group becomes SchoolGroup


snake_case (egs, school_name, age_group)
flatcase (egs, schoolname, agegroup)
camelCase (egs, schoolName, ageGroup)
PascalCase (egs, SchoolName, AgeGroup)



```
